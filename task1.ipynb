```python
import matplotlib.pyplot as pl
import json
```

```python
with open("./dataset/delhi_ncr_region.geojson",'r') as f:
    data = json.load(f)
 


for feature in data['features']:
   arr = feature['geometry']['coordinates'][0]
#    print(arr)
   x_coordinates=[]
   y_coordinates=[]   
   
   for i in arr:
    #    print(i)
       
       x_coordinates.append(i[0])
       y_coordinates.append(i[1])
           
           
   print(len(x_coordinates))
   print(len(y_coordinates))
   print("#")
   if len(x_coordinates) == len(y_coordinates):
        pl.plot(x_coordinates,y_coordinates)    
        pl.show() 
     
  
  



           
```

```python
import geopandas as gpd
import json


with open('./dataset/delhi_ncr_region.geojson') as f:
    data = json.load(f)

gdf = gpd.GeoDataFrame.from_features(data['features'])


gdf.set_crs(epsg=4326, inplace=True)

gdf.plot()
```
